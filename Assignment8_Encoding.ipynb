```python
import pandas as pd
import numpy as np
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df1.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
for col in df1.columns:
    if is_numeric_dtype(df1[col]):
        continue
    else:
        df1[col] = le.fit_transform(df1[col])
```

```python
df1.head()
```

# One Hot Encoding

```python
df2.head()
```

```python
for col in df2.columns:
    if is_numeric_dtype(df2[col]):
        continue
    else:
        df2 = pd.get_dummies(df2, columns=[col])
```

```python
df2.head()
```

# Ordinal Encoding

```python
df3.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
```

```python
for col in df3.columns:
    if not is_numeric_dtype(df3[col]):
        df3[col] = oe.fit_transform(df3[[col]])
```
